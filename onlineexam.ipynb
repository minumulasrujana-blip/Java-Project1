```python
%%writefile Onlineexam.java
import java.util.Scanner;
import java.util.Timer;
import java.util.TimerTask;
import java.io.FileWriter;
import java.io.IOException;

public class Onlineexam{

    static String[] questions = {
            "Which language is platform independent?",
            "Which keyword is used to inherit a class in Java?",
            "Which collection does NOT allow duplicates?",
            "Which company developed Java?",
            "Which method is the entry point of Java program?"
    };

    static String[][] options = {
            {"C", "C++", "Java", "Python"},
            {"this", "super", "extends", "implements"},
            {"List", "Set", "Map", "ArrayList"},
            {"Microsoft", "Sun Microsystems", "Google", "IBM"},
            {"start()", "run()", "main()", "init()"}
    };

    static int[] answers = {2, 2, 1, 1, 2};

    static int score = 0;
    static int current = 0;
    static boolean examEnded = false;

    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.println("=== ONLINE EXAMINATION SYSTEM ===");
        System.out.println("You have 60 seconds to complete the exam.\n");

        // Timer for auto submit
        Timer timer = new Timer();
        timer.schedule(new TimerTask() {
            public void run() {
                System.out.println("\nTime is up! Auto submitting exam...");
                endExam();
                System.exit(0);
            }
        }, 60000); // 60 seconds

        while (current < questions.length && !examEnded) {

            System.out.println("Q" + (current + 1) + ": " + questions[current]);

            for (int i = 0; i < 4; i++) {
                System.out.println(i + ". " + options[current][i]);
            }

            System.out.print("Your answer (0-3): ");
            int choice = sc.nextInt();

            if (choice == answers[current]) {
                score++;
            }

            current++;
            System.out.println();
        }

        timer.cancel();
        endExam();
    }

    static void endExam() {
        examEnded = true;
        saveResult();
        System.out.println("Exam Finished!");
        System.out.println("Your Score: " + score + "/" + questions.length);
    }

    static void saveResult() {
        try (FileWriter fw = new FileWriter("result.txt", true)) {
            fw.write("Score: " + score + "/" + questions.length + "\n");
        } catch (IOException e) {
            e.printStackTrace();
        }
    }
}

```
